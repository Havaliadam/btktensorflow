```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
%matplotlib inline
```

```python
yaslistesi=[10,20,30,30,30,40,50,60,70,75]
```

```python
kilolistesi=[20,60,80,85,86,87,70,90,80,95]
```

```python
numpyyaslistesi=np.array(yaslistesi)
numpykilolistesi=np.array(kilolistesi)
```

```python
numpykilolistesi
```

```python
numpyyaslistesi
```

```python
plt.plot(numpykilolistesi,numpyyaslistesi,"y")
plt.xlabel("yas")
plt.ylabel("kilo")
plt.title("kilo yaşa göre değişim grafiği")
#plt.show()
```

```python
numpydizi=np.linspace(0,10,20)
```

```python
numpydizi
```

```python
numpydizi2=numpydizi**2
```

```python
numpydizi2
```

```python
plt.plot(numpydizi,numpydizi2,"b*-")#b
```

```python
plt.subplot(1,2,1)
plt.plot(numpydizi,numpydizi2,"r*-")
plt.subplot(1,2,2)# 1 sıra olacak 2 kolon olacak 2 grafiği .izer
plt.plot(numpydizi2,numpydizi,"g--")
```

```python
benimfigure=plt.figure()
figüreaxes=benimfigure.add_axes([0.1,0.1,0.3,0.3])
figüreaxes.plot(numpydizi,numpydizi2,"g")
figüreaxes.set_xlabel("x ekseni veri isimi")
figüreaxes.set_ylabel("y ekseni veri isimi")
figüreaxes.set_title("grafik başlık")
```

```python
figure2=plt.figure()


eksen1=figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2=figure2.add_axes([0.2,0.2,0.3,0.3])

eksen1.plot(numpydizi,numpydizi2,"g")
eksen1.set_xlabel("x ekseni")
eksen1.set_ylabel("y ekseni")
eksen1.set_title("ana grafik başlık")


eksen2.plot(numpydizi2,numpydizi,"b")
eksen2.set_xlabel("x ekseni")
eksen2.set_ylabel("y ekseni")
eksen2.set_title("küçük grafik başlık")

```

```python
plt.subplots()
```

```python
(benimfigure,benimeksenler)=plt.subplots(nrows=1,ncols=2)
type(benimeksenler)

for eksen in benimeksenler:
        eksen.plot(numpydizi,numpydizi2,"b")
        eksen.set_xlabel("x ekseni")
plt.tight.layout()        
#benimeksen.plot(numpydizi,numpydizi2,"b")
#benimfigure
```

```python
yeniFigur=plt.figure()
#dpi=150
yenieksen=yeniFigur.add_axes([0.1,0.1,0.9,0.9])


yenieksen.plot(numpydizi,numpydizi2,label="numpydizisi**2")
yenieksen.plot(numpydizi,numpydizi**3,label="numpydizisi**3")
yenieksen.legend(loc=5)
                
```
