```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpydizi1=np.linspace(0,10,20)
```

```python
numpydizi1
```

```python
numpydizi2=numpydizi1**2
```

```python
numpydizi2
```

```python
(benimfigur,benimeksen)=plt.subplots()
benimeksen.plot(numpydizi1,numpydizi2,color="#3A95A8",alpha=0.2)
benimeksen.plot(numpydizi2,numpydizi1,"#C96F23")
```

```python
(yeniFigur,yenieksen)=plt.subplots()
yenieksen.plot(numpydizi1,numpydizi1+2,color="blue",linewidth=8.0)
yenieksen.plot(numpydizi1,numpydizi1+3,color="green",linewidth=1.0)

yenieksen.plot(numpydizi1,numpydizi1 + 4,color="#C96F23",linestyle="-.")
yenieksen.plot(numpydizi1,numpydizi1 + 5,color="#C96F23",linestyle="-.")

yenieksen.plot(numpydizi1,numpydizi1 + 7,color="#000000",linestyle="--",marker="o",markersize=4,markerfacecolor="red")
yenieksen.plot(numpydizi1,numpydizi1 + 8,color="#000000",linestyle="--",marker="+",markersize=4,markerfacecolor="blue")
```
